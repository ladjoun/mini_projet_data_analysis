# Appliquer les opérations de prétraitement et filtres nécessaires à ce fichier, pour ne garder que les colonnes correspondantes

```python
import pandas as pd 
df=pd.read_csv("BF.txt",sep = '\t', header = None) 
df
```

```python
#Identifiants, Noms de lieux, latitudes, longitudes
filtre_BF=df.iloc[:,[0,1,4,5]]
filtre_BF
```

```python
#Renommez les avec les noms suivants : 'ID', 'location_name', 'lat', 'long'
nouveau_noms= {0: 'ID', 1: 'location_name',4:'lat',5:'long'}
filtre_BF.rename(columns=nouveau_noms, inplace=True)
filtre_BF
```

```python
#Sauvegarder ces données dans un fichier CSV, nommez-le burkina_location.csv
filtre_BF.to_csv('burkina_location.csv', index=False)
```

# Opérations sur le fichier CSV burkina_location.csv

```python
#extraire les données contenant le nom 'gounghin', enregistrez-le sous le fichier gounghin.csv
df1=pd.read_csv("burkina_location.csv") 
dgounghin=df1.loc[df1['location_name']=='Gounghin']
dgounghin.to_csv('gounghin.csv', index=False)
```

```python
#extraire la sous-partie de la base de données (fichier burkina_location.csv), dont les noms les premières lettres des noms de lieux sont compris entre 'A' et 'P' (ordre alphabétique)
sous_df1 = df1[df1['location_name'].str[0].between('A', 'P')]
sous_df1= sous_df1.sort_values(by='location_name')
sous_df1
```

```python
#Identifiez respectivement, la latitude, la longitude minimale et les noms de lieux correspondants

# Identifier la ligne avec la latitude minimale
ligne_latitude_min = sous_df1['lat'].idxmin()

# Identifier la latitude minimale
latitude_min = sous_df1.loc[ligne_latitude_min, 'lat']

# Identifier la longitude minimale correspondante
longitude_min_correspondante = sous_df1.loc[ligne_latitude_min, 'long']

# Identifier le nom de lieu correspondant à la latitude minimale
nom_lieu_latitude_min = sous_df1.loc[ligne_latitude_min, 'location_name']

# les résultats
print("Latitude minimale:", latitude_min)
print("Longitude minimale correspondante:", longitude_min_correspondante)
print("Nom de lieu correspondant à la latitude minimale:", nom_lieu_latitude_min)
```

```python
#Quels sont les lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5)

# Filtrer les lieux selon les critères 
lieux_filtres = sous_df1[(sous_df1['lat'] >= 11) & (sous_df1['long'] <= 0.5)]
lieux_filtres
```
